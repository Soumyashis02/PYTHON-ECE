```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

fig=plt.figure()

ax=plt.axes(projection="3d")

x=[0,1,2,3,4,5,6]
y=[0,1,4,9,16,25,36]
z=[0,1,4,9,16,25,36]

ax.plot3D(x,y,z,'green')
plt.rcParams['grid.color']="violet"
ax.scatter3D(x,y,z,c=z,cmap='cividis')
plt.show()
```
