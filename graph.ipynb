```python
import numpy as np
import matplotlib.pyplot as plt
class Graph:
    def RandomArrayGraph():
      x_axis=np.random.randint(10,90,10)
      y_axis=np.random.randint(50,120,10)
      print((x_axis),type(x_axis),len(x_axis))
      print((y_axis),type(y_axis),len(y_axis))
      plt.plot(x_axis,y_axis,marker='o',mfc="red",mec="red",color="blue")
      plt.plot(x_axis,y_axis,marker='o',mfc="red",mec="red",color="yellow")
      plt.grid(color="green")
      plt.xlabel("stocks")
      plt.ylabel("share prices")
      plt.title("Stock Market")
      plt.show()
Graph.RandomArrayGraph()
```
